```python
import zipfile

with zipfile.ZipFile("Bangladeshi Currency Detection.v3-smsbs-test.yolov11.zip", 'r') as zip_ref:
    zip_ref.extractall("dataset")
```

```python
!find dataset -name "data.yaml"
```

```python
from ultralytics import YOLO

model = YOLO("yolo11n.pt")
```

```python
results = model.train(
    data="dataset/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    patience=10
)
```
